```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie pliku CSV (podmień 'ścieżka_do_pliku.csv' na własną ścieżkę)
file_path = "UEP/Salary Data (2).csv"
df = pd.read_csv(file_path)

# Obliczenie średnich zarobków w podziale na płeć
salary_by_gender = df.groupby("Gender")["Salary"].mean()

# Wykres
plt.figure(figsize=(8, 5))
salary_by_gender.plot(kind="bar", color=["blue", "pink"])
plt.xlabel("Płeć")
plt.ylabel("Średnie zarobki")
plt.title("Średnie zarobki w podziale na płeć")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Wyświetlenie wykresu
plt.show()
```
